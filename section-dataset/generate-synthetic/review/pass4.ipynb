```python
import pandas as pd

df = pd.read_parquet('synthetic-biased-job-descriptions.parquet')
print(f'Rows: {df.shape[0]}')
df.head(10)
```

```python
label_columns = df.columns[df.columns.str.startswith('label_')]
label_columns
```

```python
def print_samples(category, n=5):
    sample_df = df[df[f'label_{category}'] == True].sample(n)

    for idx, row in sample_df.iterrows():
        print(f'#### SAMPLE {idx} ####')
        for column in label_columns:
            print(f'[{column}={row[column]}]', end='\n')
        print(f'----------------------\n{row["text"]}')
```

```python
print_samples('age',n=5)
```
